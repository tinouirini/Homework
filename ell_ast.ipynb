```python
declared_males = [215, 175, 178, 155, 194, 175, 180, 196, 180, 150, 136, 246]
declared_females = [87, 69, 142, 42, 57, 60, 53, 54, 45, 49, 52, 36]
found_males = [64, 54, 65, 58, 76, 66, 70, 77, 71, 65, 40, 57]
found_females = [79, 57, 53, 37, 48, 45, 39, 40, 37, 41, 38, 22]

# Δημιουργία πίτας για τις δηλώσεις εξαφανίσεων ανά φύλο
plt.figure(figsize=(10, 6))
plt.pie([sum(declared_males), sum(declared_females)], labels=['Άρρενες', 'Θήλεις'], autopct='%1.1f%%', startangle=140)
plt.title('Δηλώσεις Εξαφανίσεων ανά φύλο')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python

# Δημιουργία πίτας για τους εντοπισμούς εξαφανισθέντων ανά φύλο
plt.figure(figsize=(10, 6))
plt.pie([sum(found_males), sum(found_females)], labels=['Άρρενες', 'Θήλεις'], autopct='%1.1f%%', startangle=140)
plt.title('Εντοπισμοί Εξαφανισθέντων ανά φύλο')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Δημιουργία γραφήματος πίτας για τις δηλώσεις εξαφανίσεων ανά μήνα και φύλο
month = [
    'Ιανουάριος', 'Φεβρουάριος', 'Μάρτιος', 'Απρίλιος', 'Μάϊος', 'Ιούνιος',
    'Ιούλιος', 'Αύγουστος', 'Σεπτέμβριος', 'Οκτώμβριος', 'Νοέμβριος', 'Δεκέμβριος'
]

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(declared_males, labels=month, autopct='%1.1f%%', startangle=140)
plt.title('Δηλώσεις Εξαφανίσεων Αρρένων ανά Μήνα')
plt.axis('equal')

plt.subplot(1, 2, 2)
plt.pie(declared_females, labels=month, autopct='%1.1f%%', startangle=140)
plt.title('Δηλώσεις Εξαφανίσεων Θηλέων ανά Μήνα')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Δημιουργία γραφήματος πίτας για τους εντοπισμούς εξαφανισθέντων ανά μήνα και φύλο
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(found_males, labels=month, autopct='%1.1f%%', startangle=140)
plt.title('Εντοπισμοί Εξαφανισθέντων Αρρένων ανά Μήνα')
plt.axis('equal')
plt.subplot(1, 2, 2)
plt.pie(found_females, labels=month, autopct='%1.1f%%', startangle=140)
plt.title('Εντοπισμοί Εξαφανισθέντων Θηλέων ανά Μήνα')
plt.axis('equal')

plt.tight_layout()
plt.show()
```
